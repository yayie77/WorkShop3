***What percent of the total number of passengers on the Titanic belonged to each age group(e.g., child, adult, senior)?***

To answer the question perform following operation:

1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age 18{64), and "senior" (age 65 and above).

```python
def categorize_age(age):
  if age<18:
    return 'child'
  elif age>=18 and age<64:
    return 'adult';
  else:
      return 'senior'
df['Age group']=df['Age'].apply(categorize_age)
df.head()
```

2. Calculate the total number of passengers on the Titanic.

```python
total_num_of_passengers=df['PassengerId'].count()
print(f"Total number of passengers on the Titanic is {total_num_of_passengers}.")
```

3. Count the number of passengers in each age group.

```python
age_group_counts=df['Age group'].value_counts()
print(f"The number of passengers in each age group:{age_group_counts}")
```

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
age_total_proportion=age_group_counts/total_num_of_passengers
print(f"The proportion of passengers in each age group: {age_total_proportion}")
```

5. Display the proportion as a percentage.

```python
age_group_percentage=age_total_proportion*100
print(f"The proportion as a percentage:{age_group_percentage}.")
```
